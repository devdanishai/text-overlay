```python
# Install required packages
#!pip install rembg[gpu] onnxruntime-gpu

# Import libraries
from rembg import remove
from PIL import Image
from google.colab import drive
from google.colab import files
import os

# Mount Google Drive
drive.mount('/content/drive')

# Create paths
input_folder = '/content/drive/My Drive/background_removal'
output_folder = '/content/drive/My Drive/background_removal/output'

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Option 1: Process image from Drive
def process_from_drive(image_name):
    input_path = os.path.join(input_folder, image_name)
    # Ensure output file has .png extension
    output_name = os.path.splitext(image_name)[0] + '.png'
    output_path = os.path.join(output_folder, 'output_' + output_name)

    input_image = Image.open(input_path)
    output_image = remove(input_image)
    output_image.save(output_path, format='PNG')  # Explicitly specify PNG format
    print(f"Saved to: {output_path}")

# Option 2: Process uploaded image
def process_uploaded_image():
    print("Please select an image to upload:")
    uploaded = files.upload()
    filename = next(iter(uploaded))

    input_image = Image.open(filename)
    output_image = remove(input_image)

    # Ensure output file has .png extension
    output_name = os.path.splitext(filename)[0] + '.png'
    output_path = os.path.join(output_folder, 'output_' + output_name)
    output_image.save(output_path, format='PNG')  # Explicitly specify PNG format
    print(f"Saved to: {output_path}")

    # Optional: Download the result directly
    files.download(output_path)

# Choose your preferred method:
# 1. For image in Drive (replace 'your_image.jpg' with your image name):
# process_from_drive('your_image.jpg')

# 2. For uploading new image:
process_uploaded_image()
```
